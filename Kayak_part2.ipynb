```python
# Import os => Library used to easily manipulate operating systems
## More info => https://docs.python.org/3/library/os.html
import os 

# Import logging => Library used for logs manipulation 
## More info => https://docs.python.org/3/library/logging.html
import logging

# Import scrapy and scrapy.crawler 
import scrapy
from scrapy.crawler import CrawlerProcess
import pandas as pd
```

```python
class hotelsSpider(scrapy.Spider):
    # Name of your spider
    name = "hotels"

    # Starting URL
    start_urls = ['https://www.booking.com/index.html']

    # Parse function for login
    def parse(self, response):
        # FormRequest used to login
        return scrapy.FormRequest.from_response(
            response,
            formdata={'ss': destination},
            callback=self.after_search
        )

    # Callback used after login
    def after_search(self, response):
        hotels = response.css('.sr_item')
        i = 0
        for hotel in hotels:
            yield {
                'name': hotel.css('.sr-hotel__name::text').get(),
                'url': "https://www.booking.com" + hotel.css('.hotel_name_link').attrib["href"],
                'coords': hotel.css('.sr_card_address_line a').attrib["data-coords"],
                'score': hotel.css('.bui-review-score__badge::text').get(),
                'description': hotel.css('.hotel_desc::text').get()
            }
```

```python
data_1 = pd.read_csv('classement_météo.csv')
list_cities = list(data_1.loc[:,'display_name'])
```

```python

filename = "hotels_" + list_cities[0].replace(" ", "-") + ".json"

if filename in os.listdir('src/'):
        os.remove('src/' + filename)

process = CrawlerProcess(settings = {
    'USER_AGENT': 'Chrome/84.0 (compatible; MSIE 7.0; Windows NT 5.1)',
    'LOG_LEVEL': logging.INFO,
    "FEEDS": {
        'src/' + filename: {"format": "json"},
    }
})

process.crawl(hotelsSpider)
process.start()
```
